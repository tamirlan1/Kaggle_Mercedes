```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
train_df.shape, test_df.shape 
```

Remove columns with only one unique value

```python
one_unique_value_cols = ['X11', 'X93', 'X107', 'X233', 'X235', 'X268', 'X289', 'X290', 'X293', 'X297', 'X330', 'X347']
for col in one_unique_value_cols:
    print col, sum(train_df[col]), sum(test_df[col])
```

```python
train_df.drop(one_unique_value_cols, axis=1, inplace=True)
test_df.drop(one_unique_value_cols, axis=1, inplace=True)
train_df.shape, test_df.shape 
```

```python
thr = 10
rare_cols = []
for col in train_df.columns:
    if col in ["ID", "y", "X0", "X1", "X2", "X3", "X4", "X5", "X6", "X8"]:
        continue
    if sum(train_df[col]) < thr and sum(test_df[col]) < thr:
        rare_cols.append(col)
        print col, sum(train_df[col]), sum(test_df[col])
rare_cols, len(rare_cols)
```

```python
train_df.drop(rare_cols, axis=1, inplace=True)
test_df.drop(rare_cols, axis=1, inplace=True)
train_df.shape, test_df.shape 
```

```python
train_df.var(axis=0).reset_index().sort(ascending=False)
```

```python
train_df.to_csv('clean_train.csv')
test_df.to_csv('clean_test.csv')
```
